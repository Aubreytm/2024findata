```python
A_2356 = (df1.loc['2356 Inventec','Total current assets'] - df1.loc['2356 Inventec','Total current liabilities'])/df1.loc['2356 Inventec','Total assets']
```

```python
X1 ='Total current assets'
X2 ='Total current liabilities'
X3 ='Total assets'
X4 ='Total retained earnings'
X5 ='Earnings before interest and tax'
X6 ='Total equity'
X7 ='Total liabilities'
X8 ='Total operating revenue'
Inventec ='2356 Inventec'
Asustek ='2357 Asustek'
```

```python
A_Inventec = (df1.loc[Inventec,X1] - df1.loc[Inventec,X2])/df1.loc[Inventec, X3]
```

```python
B_Inventec = df1.loc[Inventec,X4]/df1.loc[Inventec, X3]
```

```python
C_Inventec = df2.loc[Inventec,X5]/df1.loc[Inventec, X3]
```

```python
D_Inventec = df1.loc[Inventec,X6]/df1.loc[Inventec, X7] #調整分子項定義，採用股東權益帳面價值
```

```python
E_Inventec = df2.loc[Inventec,X8]/df1.loc[Inventec, X3]
```

```python
Z_score_Inventec = 1.2*A_Inventec + 1.4*B_Inventec + 3.3*C_Inventec + 0.6*D_Inventec +1.0*E_Inventec
```

```python
Z_score_Inventec, A_Inventec, B_Inventec, C_Inventec, D_Inventec, E_Inventec
```

```python
A_Asustek = (df1.loc[Asustek,X1] - df1.loc[Asustek,X2])/df1.loc[Asustek, X3]
```
